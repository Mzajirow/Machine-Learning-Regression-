```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\USER\Documents\Ejiro\Hamoye AI Lab\energydata_complete.csv")
df.head(5)
```

```python
X = df['T2'].values.reshape(-1, 1)
Y = df['T6'].values.
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X, Y)
```

```python
predicted_values = linear_model.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(Y, predicted_values)
round(r2_score, 2)
```

```python
features_df = df.drop(columns = ['date', 'lights'])
features_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features_df)
```

```python
target = features_df["Appliances"]
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(features_scaled, target, test_size=0.7, random_state=42)
```

```python
linear_model.fit(X_train, Y_train)
```

```python
predicted_values = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(Y_test, predicted_values)
round(MSE, 2)
```

```python
RSS = np.sum((Y_test - predicted_values)**2)
round(RSS, 2)
```
